```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('titanic_train.csv')
train['Pclass'].unique()
```

#### label encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train['Encoded Pclass'] = le.fit_transform(train['Pclass'])
train['Encoded Pclass'].head()
```

```python
print(le.classes_)
```

```python
train.select_dtypes(include='object').columns
```

#### one-hot encoding

```python
train['Sex'].unique()
```

```python
train = pd.get_dummies(train, columns= ['Sex'], drop_first=True)
train.head()
```
