#### handling NAN
Age         177 - fill na with median (not mean because outliers may exist)
Cabin       687 - drop the coloumn because 687/891 is a big number
Embarked      2 - fill with mode

```python
train['Age'] = train['Age'].fillna(train['Age'].median())
train['Age'].isna().sum()
```

```python
train['Embarked'] = train['Embarked'].fillna(train['Embarked'].mode()[0])
train['Embarked'].isna().sum()
```
