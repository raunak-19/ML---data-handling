```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('titanic_train.csv')
```

```python
train['Age']
```

```python
train['Age'] = train['Age'].fillna(train['Age'].median())
```

```python
sns.boxplot(x=train['Age'])
```

```python
outliers = []
q1 = train['Age'].quantile(0.25)
q3 = train['Age'].quantile(0.75)
IQR = q3-q1
for i in train['Age']:
    if i < q1 - 1.5*(IQR) or i > q3 + 1.5*(IQR) :
        outliers.append(i)
len(outliers)
```

```python
lower = q1 - 1.5*(IQR)
upper = q3 + 1.5*(IQR)
train2 = train[
    (train['Age'] >= lower) & 
    (train['Age'] <= upper)
]
train2
```

```python
sns.boxplot(x=train2['Age'])
```
