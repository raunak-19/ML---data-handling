```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

train = pd.read_csv("titanic_train.csv")
```

```python
train.select_dtypes(include='number').head()
```

```python
corr_matrix = train.corr(numeric_only=True) #correlation matrix
corr_matrix
```

1. Pclass has a moderate negative correlation with survival.

2. Fare has a positive correlation with survival.

3. Age has a weak relationship.

4. Correlation does not imply causation.

```python
sns.heatmap(corr_matrix,annot=True,cmap = 'coolwarm')
plt.show()
```

suppose our target is 'Survived'

```python
corr_matrix['Survived'].sort_values(ascending=False)
```

we see Pclass has a strong correlation with survived


when fare increases class decreases(1<2<3)
