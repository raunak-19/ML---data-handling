```python
import numpy as np
import pandas as pd
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv('covid_toy.csv')
train.sample(5)
```

```python
transformer = ColumnTransformer([
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3', OneHotEncoder(sparse_output=False,drop='first'),['gender','city']),
    ('tnf4', SimpleImputer(),['fever']),
], remainder='passthrough')
```

```python
x_train,x_test,y_train,y_test = train_test_split(train.drop(columns=['has_covid']), train['has_covid'],test_size=0.2, random_state=42,stratify=train['has_covid'] )
```

```python
x_train_transformed = transformer.fit_transform(x_train)
x_test_transformed = transformer.transform(x_test)
```
